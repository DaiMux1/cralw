```python
def get_documents(url):
    document = {
        'topic': '',
        'time': '',
        'vender': '',
        'header': '',
        'body': '',
        'url': url
    }
    req = requests.get(url, headers=agent)
    soup = BeautifulSoup(req.text, 'html.parser')
#     print(soup)
    # get topic 
    try:
        topics = []
        ol = soup.find('ol', attrs={'class': 'l-topicsPath'})
        
        lis = ol.find_all('li', attrs={'class': 'l-topicsPath_item'})
        li1 = lis[0:len(lis)-1]
        
        for li in li1:
            a = li.find('a')
            if a is not None:
                topics.append(a.text.strip())
#         print(topics)
        document['topic'] = ''.join(topics)  
        #get header
        li2 = lis[-1]
        a = li2.find('a')
        document['header'] = a.text.strip()
        
    except Exception as e:
        print(e)
        
    # get vender
    try:
        author = []
        mata = soup.find('div', attrs={'class': 'p-article_header_author'})
        address = mata.find('address', attrs={'class': 'p-article_header_author_fullName'})
        a = address.find('a')
        document['vender'] = ''.join(a.text.strip())
    
    except:
        print(url)
        
    try:
        document['time'] = soup.find('time', attrs={'class': 'p-article_header_date'})['datetime']
    
    except:
        print(url)
        
    try:
        article = soup.find('div', attrs={'class': 'articleBody p-article_body'})
        text = ''
        links = article.find_all('p')
        for link in links:
            text += link.text.strip()
        document['body'] = text
        
    except:
        print(url)
        
    return document
```

```python
from tqdm import tqdm
import pandas as pd

i = 0

for name in dirnames:
    print('===========')
    print(name)
    paths = os.listdir(os.path.join(out_dir, name))
#     print(paths)
    txt_paths = [path for path in paths if os.path.isdir(os.path.join(out_dir, name, path))]
    
    print(txt_paths)
    
    if len(txt_paths) == 0:
        continue
        
    
    
    for c_path in txt_paths:
#         print('CHILD', c_path)
        c_a_paths = os.listdir(os.path.join(out_dir, name, c_path))
#         print(c_a_paths)
        for path in c_a_paths:
#             print(path)
            
            with open(os.path.join(out_dir, name, c_path, path), 'r') as f:
                article_urls = f.read().splitlines()
                
            for a_url in tqdm(article_urls):
                name_csv = a_url.split('/')[-2]
#                 print(name_csv)
                data = pd.DataFrame(columns=['topic', 'time', 'vender', 'header', 'body', 'url'])
                document = get_documents(a_url)
                data = data.append(document, ignore_index=True)
                data.to_csv(os.path.join(data_dir, name_csv + '.csv'))
    
#     data.to_csv(os.path.join(data_dir, name + '.csv'))
    
            
        
    
    
        
    
    
    
    
```
